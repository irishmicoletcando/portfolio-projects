## 5. Explore Lego Sets

<p>Let us use this data to explore how the average number of parts in Lego sets has varied over the years.</p>

### Task 5: Instructions
- Read the data in `datasets/sets.csv` as a DataFrame named `sets`.
- Create a summary of the average number of parts per year and save it as `parts_by_year`.
- Plot the average number of parts per year.

```python
import matplotlib.pyplot as plt
# Read sets data as `sets`
sets = pd.read_csv('datasets/sets.csv')

# Create a summary of average number of parts by year: `parts_by_year`
parts_by_year = round(sets.groupby('year')['num_parts'].sum() / sets.groupby('year')['num_parts'].count(), 2)
print(parts_by_year)

# Plot trends in average number of parts by year
plt.figure(figsize=(15, 6))
parts_by_year.plot(kind='line', x='year', y='num_parts', marker='o')
plt.show()
```
